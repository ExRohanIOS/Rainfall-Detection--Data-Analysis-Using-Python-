```python
import numpy as py
import pandas as pd
import sklearn as sk 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("austin_weather_final.csv")
```

```python
data
```

```python
X = data.drop(['PrecipitationSumInches'],axis=1)
```

```python
Y = data["PrecipitationSumInches"]
```

```python
Y.head()
```

```python
Y= Y.values.reshape(-1,1)
```

```python
Y
```

```python
index_days = 798
days = [i for i in range(Y.size)]
```

```python
clf = LinearRegression()
clf.fit(X,Y)
```

```python
plt.title('The Precipictation Of Trend Graph')
plt.scatter(days,Y,color='g')
plt.scatter(days[index_days],Y[index_days],color='b')
plt.xlabel("Days")
plt.ylabel("Precipitation Level In Inches")
plt.show()
```

```python
vis=X.filter(['TempAvgF','DewPointAvgF','HumidityAvgPercent','SeaLevelPressureAvgInches','VisibilityAvgMiles','WindAvgMPH'])
```

```python
vis
```

```python
print('The Precipictation vs Attributes Of Trend Graph')
for i in range(vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days,vis[vis.columns.values[i][:100]],color='g')
    plt.scatter(days[index_days],vis[vis.columns.values[i]][index_days],color='b')
    plt.title(vis.columns.values[i])
    
plt.show()
```
